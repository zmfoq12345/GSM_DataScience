```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# 데이터 생성
X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([0,1,1,0])
```

```python
# 2차원, 데이터 개수 4개
X
```

```python
# 2차원, 데이터 개수 4개
y
```

```python
# 신경망 모델 생성
model = Sequential()

# 은닉층
model.add(Dense(units = 32, # 신경망의 개수
                input_dim=2,# 입력이 들어오는 차원 개수 즉, X의 column개수
                activation="sigmoid")) # 활성화 함수 : relu, sigmoid, tanh
# 출력층
model.add(Dense(units = 1, # 신경망의 개수 (이진분류 문제는 출력층이 1개로 지정됨)
                activation="sigmoid")) # 이진분류의 출력층 활성화 함수는 sigmoid
```

```python
# 신경망 구조 파악
model.summary()
```

```python
# error(loss)와 optimizer 작성
model.compile(loss="binary_crossentropy", # 이진분류는 binary_crossentropy
             optimizer="adam", # RMSPro
             metrics=["accuracy"]) # 평가
```

```python
model.fit(X, # 문제
          y, # 답
          epochs=3000, # 학습 횟수
          verbose=2) # 출력창 제어 (0: 숨김, 1: 세밀하게, 2: 간단하게)
```

```python
# 평가 (loss(오차), accuracy(정확도))
evalu = model.evaluate(X, y, verbose=0) # 문제, 답, 출력창 제어
print("loss :", evalu[0])
print("accuracy :", evalu[1])
```

```python
# 에측
model.predict(X)
# 결과는 츨력층의 sigmoid함수로 인해 0~1 사이의 값이 나옴
# 0.5보다 큰 경우 1로 분류
# 0.5보다 작은 경우 0으로 분류
```

```python
# 예측
model.predict_classes(X)
```

```python
# 가중치 및 편향 확인
model.get_weights()
```

## 다중분류

```python
X = np.array([[0,0], [0,1], [1,0], [1,1]])
# one-hot encoding된 결과
y = np.array([[1,0], [0,1], [0,1], [1,0]])
# 
X
```

```python
y
```

```python
model = Sequential()
model.add(Dense(units=32, 
                input_dim=2, 
                activation="sigmoid"))
# 출력층
model.add(Dense(units=2, # y열의 개수
                activation="softmax")) # 다중분류 활성화 함수 softmax 국룰
```

```python
model.compile(loss="categorical_crossentropy", # 다중분류 categorical_crossentropy
             optimizer="RMSprop",
             metrics=["accuracy"])
```

```python
model.summary()
```

```python
model.fit(X, y, epochs=2000)
```

```python
model.evaluate(X, y, verbose=0)
```

```python
# 숫자가 더 높은 열이 정답
model.predict(X)
```

```python
# 예측
model.predict_classes(X)
```
