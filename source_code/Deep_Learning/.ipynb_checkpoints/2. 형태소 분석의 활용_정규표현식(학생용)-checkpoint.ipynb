# 정규표현식 사용하기 연습

```python
import re
```

#### re.compile('정규표현식') : 찾고자 하는 패턴이 빈번할 경우 정규표현식을 컴파일하여 사용
#### re.sub(exp, rep, text) : 정규표현식 exp에 매칭되는 문자와 text 내의 매칭되는 문자를 rep로 대체한다.
#### re.split(exp, text) : 정규표현식 exp를 기준으로 text를 분리하여 리스트로 반환

```python
import re
text = "ha 이번 방학엔 '사이코지만 괜찮아' 드라마 정주행 할거야ㅋㅋ지금은 너무 바빠서 시간이 없어ㅠㅠ"

obj_word = re.compile('[ㄱ-ㅎㅏ-ㅣ가-힣]')
sub_word = '<SUB>'

result = re.sub(obj_word, sub_word, text)

# split_result = re.split(" ", text)

print(f"[원래문장] {text}")
print(f"'{sub_word}'로 대체된 결과 : {result}")

# print(f"공백을 기준으로 text 분리 : {split_result}")
```

```python
import re

text = "ha 이번 방학엔 '사이코지만 괜찮아' 드라마 정주행 할거야ㅋㅋ지금은 너무 바빠서 시간이 없어ㅠㅠ"

obj_word = re.compile('[a-zA-Z]')
sub_word = '<SUB>'

result = re.sub(obj_word, sub_word, text)
                
print(f"[원래문장] {text}")
print(f"'{sub_word}'로 대체된 결과 : {result}")
```

###  "I love you. 사랑해." 문장을 영어와 한글 각각 따로 출력하기

```python
import re 

text = "I love you. 사랑해."

eng = re.compile('[^ㄱ-ㅎㅏ-ㅣ가-힣 ]')
kor = re.compile('[^a-zA-Z ]')

result1 = re.sub(eng, "", text)
result2 = re.sub(kor, "", text)

print(f"한글만 출력(공백 포함) : {result1}")
print(f"영어만 출력(공백 포함) : {result2}")
```

```python
import re 

text = "I love you. 사랑해."

eng = re.compile('[^가-힣]') #한글을 제외한 모든 것과 매칭
kor = re.compile('[^A-Za-z ]') #영어, 공백을 제외한 모든 것과 매칭

result1 = re.sub(eng, "", text)
result2 = re.sub(kor, "", text)

print(f"한글만 출력(공백 포함) : {result1}")
print(f"영어만 출력(공백 포함) : {result2}")
```

###  이메일 주소와 핸드폰 번호를 공백("")으로 바꾸기

```python
import re 

text = "제 이메일 주소는 abc_1-23@gsm.hs.kr입니다. 번호는 010-1234-5678입니다."

compiling = re.compile('[a-zA-Z0-9\-_]+@+[a-z\.]+|[0-9]+-[0-9]+-[0-9]+')

result = re.sub(compiling, "", text)

print(f"변경한 문장 : {result}")
```

```python
import re 

text = "제 이메일 주소는 abc_1-23@gsm.hs.kr입니다. 번호는 010-1234-5678입니다."

compiling = re.compile('([a-zA-Z0-9\-_]+@+[a-zA-Z\.]+)|([0-9\-]{3,14})')

result = re.sub(compiling, "", text)

print(f"변경한 문장 : {result}")
```

## 정규표현식 사용하여 '운수 좋은 날 '정제하기

```python
text = "아니,\n이 행운을 놓칠 수 없다 하\n였다.\n“일 원 오십 전은 너무 과한데.”\n이런 말을 하며 학생은 고개를 기웃하였다.\n"

com = re.compile('(?<=[가-힣,])\n(?=[가-힣])')

result = re.sub(com, "", text)
result
```

#### (?<=[가-힣])\n : 한글 다음에 \n이 올 경우 \n와 매칭
#### \n(?=[가-힣]) : \n 다음에 한글이 올 경우 \n와 매칭

```python
import re

def refine(contents):
    new_line = re.compile('(?<=[가-힣,])\n(?=[가-힣])')
    
    contents = re.sub(new_line, "", contents)
    return contents
```

```python
luckday = refine(luckday)
luckday
```
