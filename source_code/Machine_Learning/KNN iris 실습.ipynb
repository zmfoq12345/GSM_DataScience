# 꽃받침(Sepal) 데이터만 활용하여 분석

```python
X2 = X[:,:2]
X2[:3]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = \
    train_test_split(X2, # 문제
                     y,  #답
                     random_state=3,# 랜덤으로 나누되 항상 똑같이 나뉨
                     test_size=0.2, # test가 20%로 나눔
                     stratify=y)    
"""random_state, test_size, stratify 순서로 설명"""
```

```python
np.bincount(y_train), np.bincount(y_test)
```

```python
"""데이터 시각화"""
markers = ['o','+','^']
for i in range(3):
    xs = X_train[:,0][y_train==i]
    ys = X_train[:,1][y_train==i]
    plt.scatter(xs,ys,marker=markers[i])
plt.legend(iris.target_names)
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors = 3)
model.fit(X_train, y_train)
print("훈련 점수 : {}".format(model.score(X_train, y_train)))
print("테스트 점수 : {}".format(model.score(X_test, y_test)))
```

## 점수가 낮으니 setosa와 setosa가 아닌걸로 이진 분류

```python
y2 = y.copy()
y2[y2==2] = 1
```

```python
X_train, X_test, y_train, y_test = \
    train_test_split(X2, y2, random_state=3, test_size=0.2, stratify=y)
```

```python
"""데이터 시각화"""
markers = ['o','+','^']
for i in range(3):
    xs = X_train[:,0][y_train==i]
    ys = X_train[:,1][y_train==i]
    plt.scatter(xs,ys,marker=markers[i])
plt.legend(iris.target_names)
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")
```

```python
model = KNeighborsClassifier()
model.fit(X_train, y_train)
print("훈련 점수 : {}".format(model.score(X_train, y_train)))
print("테스트 점수 : {}".format(model.score(X_test, y_test)))
```

# 실습(꽃잎으로 분류 모델 만들기)

```python
X3 = X[:, 2:]
```

```python
X_train, X_test, y_train, y_test = \
    train_test_split(X3, # 문제
                     y,  #답
                     random_state=3,# 랜덤으로 나누되 항상 똑같이 나뉨
                     test_size=0.2, # test가 20%로 나눔
                     stratify=y)    
```
