```python
markers = ['^', 's', 'o']
for idx, mk in enumerate(markers):
    x_axis_data = iris_Df[iris_Df['target']==idx]['sepal length (cm)']
    y_axis_data = iris_Df[iris_Df['target']==idx]['sepal width (cm)']
    plt.scatter(x_axis_data, y_axis_data, marker = mk, label = iris.target_names[idx])

plt.legend()
plt.xlabel('sepal length')
plt.xlabel('sepal width')
plt.show()
```

```python
iris_Df2= iris_Df.copy()

del iris_Df2['target']
iris_Df2.head(3)
```

```python
iris_Df3 = iris_Df.copy()

iris_Df4 = iris_Df3.drop('target', axis = 1, inplace = True)
iris_Df3.head()
```

```python
iris_Df4.head()
```

```python
from sklearn.preprocessing import StandardScaler

iris_scaled = StandardScaler().fit_transform(iris_Df2)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 2)

pca.fit(iris_scaled)
print(pca.components_)

iris_pca = pca.transform(iris_scaled)
print(iris_pca.shape)
```

```python
iris.pca_df = pd.DataFrame(pca.transform(iris_scaled), columns = ['pca1', 'pca2'])
iris.pca_df.head()
```

```python
iris.pca_df['target'] = iris.target
iris.pca_df.head()
```

```python
markers = ['^', 's', 'o']
for idx, mk in enumerate(markers):
    x_axis_data = iris.pca_df[iris.pca_df['target']==idx]['pca1']
    y_axis_data = iris.pca_df[iris.pca_df['target']==idx]['pca2']
    plt.scatter(x_axis_data, y_axis_data, marker = mk, label = iris.target_names[idx])

plt.legend()
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.grid()
plt.show()
```
