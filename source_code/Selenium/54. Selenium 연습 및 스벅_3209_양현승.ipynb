```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup as bs
import pandas as pd
import time
```

```python
driver = wb.Chrome()

url = 'http://www.istarbucks.co.kr/store/store_map.do'
driver.get(url)
```

```python
# alert = driver.switch_to_alert()
# alert.accept()
```

```python
driver.page_source

time.sleep(1)
driver.find_element_by_css_selector('header.loca_search > h3 > a').click()
time.sleep(1)
driver.find_element_by_css_selector('ul.sido_arae_box > li + li + li').click()
time.sleep(1)
driver.find_element_by_css_selector('ul.gugun_arae_box > li').click()
```

```python
soup = bs(driver.page_source, 'html.parser')

name = [ i.text for i in soup.select('#mCSB_3_container > ul > li > strong')]
li = [ i.text for i in soup.select('#mCSB_3_container > ul > li > p')]
addr = [x[:-9] for x in li]
tel = [x[-9:] for x in li]

for n, a, t in zip(name, addr, tel):
    print('지점명:', n)
    print('주소:', a)
    print('번호:', t)
    print('---------------------------------------------')
```

```python
names = []
addrs = []
tels = []
for i in range(len(name)):
    names.append(name[i])
    addrs.append(addr[i])
    tels.append(tel[i])

    data = {'name':names, 'addr':addrs, 'tel': tels}
df = pd.DataFrame(data)
df.head()
```

```python
df.to_csv("200528 스타벅스 매장 정보.csv", encoding='uck-kr', index = False)
```
