```python
from selenium import webdriver as wb
from bs4 import BeautifulSoup as bs
import time
import pandas as pd
```

```python
url ="https://www.hsd.co.kr/menu/menu_list"
driver = wb.Chrome()
driver.get(url)
driver.page_source

# [더보기] 버튼을 클릭해서 전체페이지 내용이 출력되도록 함.
for i in range(3):
    driver.find_element_by_css_selector('a.c_05').click()
    time.sleep(2)
    
soup = bs(driver.page_source,'html.parser')
menu= soup.select('h4.h')
price = soup.select('div.item-price > strong')

for i in range(len(menu)):
    print('메뉴: ' + menu[i].text)
    print('가격: ' + price[i].text)
    print('-'*40)
```

```python
menus =[]
prices = []
# for i in range(len(menu)):
#     menus.append(menu[i].text)
#     prices.append(price[i].text)

for m, p in zip(menu, price):
    menus.append(m.text)
    prices.append(p.text)
```

```python
data = {'menu': menus, 'price': prices}
df = pd.DataFrame(data)
df.head()
```

```python
df.to_csv('hsdData.csv', encoding='euc-kr', index=False)
```
